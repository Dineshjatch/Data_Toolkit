# Theory Questions
1. What is NumPy, and why is it widely used in Python?
- NumPy is a Python library used for numerical computations. It provides fast, efficient array operations, supports multidimensional arrays, and integrates well with other data science libraries.

2. How does broadcasting work in NumPy?
- Broadcasting allows NumPy to perform operations on arrays of different shapes. It stretches the smaller array to match the shape of the larger one without making extra copies in memory.

3. What is a Pandas DataFrame?
- A DataFrame is a two-dimensional labeled data structure in Pandas. It is similar to a table in Excel or a SQL table, where each column can have a different data type.

4. Explain the use of the groupby() method in Pandas.
- The groupby() method splits data into groups based on some criteria. It allows aggregation like sum, mean, or count for each group, helping in data analysis.

5. Why is Seaborn preferred for statistical visualizations?
- Seaborn provides high-level functions for attractive and informative statistical graphics. It integrates well with Pandas and automatically handles many plotting details.

6. What are the differences between NumPy arrays and Python lists?
- NumPy arrays are faster and consume less memory than Python lists. They support vectorized operations, while lists require loops for element-wise calculations.

7. What is a heatmap, and when should it be used?
- A heatmap is a graphical representation of data where values are shown using color. It’s best used to visualize correlation matrices or large datasets to spot patterns.

8. What does the term “vectorized operation” mean in NumPy?
- Vectorized operations allow operations on entire arrays without explicit loops. They are faster and more efficient, using low-level optimized code behind the scenes.

9. How does Matplotlib differ from Plotly?
- Matplotlib is a static plotting library ideal for basic plots. Plotly is interactive, offering zooming, hovering, and clickable features, which are useful for dashboards.

10. What is the significance of hierarchical indexing in Pandas?
- Hierarchical indexing (MultiIndex) allows more complex data analysis by using multiple index levels. It helps in working with higher-dimensional data in a 2D format.

11. What is the role of Seaborn’s pairplot() function?
- pairplot() creates scatter plots for each pair of numerical features and histograms for individual features. It helps in visualizing relationships in a dataset.

12. What is the purpose of the describe() function in Pandas?
- The describe() function provides statistical summaries like mean, median, min, max, and quartiles. It is useful for quickly understanding numerical data.

13. Why is handling missing data important in Pandas?
- Missing data can lead to incorrect analysis or errors. Pandas provides tools like fillna() and dropna() to clean or handle missing values appropriately.

14. What are the benefits of using Plotly for data visualization?
- Plotly creates interactive, web-based plots that support zooming, hovering, and tooltips. It’s ideal for dashboards and presentations.

15. How does NumPy handle multidimensional arrays?
- NumPy supports arrays with any number of dimensions using the ndarray object. It provides functions to reshape, slice, and perform operations across axes.

16. What is the role of Bokeh in data visualization?
- Bokeh is a Python library for creating interactive plots and dashboards in web browsers. It is useful for real-time streaming data and interactive applications.

17. Explain the difference between apply() and map() in Pandas.
- map() is used for element-wise operations on a Series. apply() is more flexible and can be used on both Series and DataFrames, applying functions row- or column-wise.

18. What are some advanced features of NumPy?
- Advanced features include broadcasting, vectorized operations, masked arrays, linear algebra functions, and integration with C/C++ for performance.

19. How does Pandas simplify time series analysis?
- Pandas has built-in support for date/time data, resampling, time shifting, and frequency conversion. This simplifies operations like moving averages and rolling windows.

20. What is the role of a pivot table in Pandas?
- A pivot table summarizes data by transforming rows into columns. It helps in grouping and aggregating large datasets for easier analysis.

21. Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy slicing returns views instead of copies, and it’s implemented in C for performance. This makes it significantly faster than Python list slicing.

22. What are some common use cases for Seaborn?
- Seaborn is used for visualizing distributions, correlations, regression lines, and statistical relationships. It’s popular in EDA and presentation-quality plotting.

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame
import numpy as np
import pandas as pd
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
mean_A = df['A'].mean()
print(mean_A)
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import pandas as pd
df = sns.load_dataset('iris')
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#5. Generate a bar plot using Plotly
import plotly.express as px
import pandas as pd

data = {
    "Category": ["A", "B", "C", "D"],
    "Values": [23, 45, 12, 67]
}

df = pd.DataFrame(data)

fig = px.bar(df, x="Category", y="Values", title="Simple Bar Plot")
fig.show()
```

```python
#9. A Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}
df=pd.DataFrame(data)
threshold=30
filtered_df=df[df['Age']>threshold]
print(filtered_df)
```
